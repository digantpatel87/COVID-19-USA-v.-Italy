```python
import pandas as pd
import numpy as np
import requests
import os
import json
import datetime as dt
from dateutil.parser import parse
```

```python
#json to dataframe
us_df = pd.read_json('data/us_covid19.json')
us_df['date']=us_df['date'].astype(str)
```

```python
#change date format
for i, row in us_df.iterrows():
    parsed= parse(row.date)
    us_df.at[i,'date']=parsed.strftime('%Y-%m-%d')
```

```python
us_df.head()
```

```python
state_df = pd.read_json('data/us_states.json')
state_df['date']=state_df['date'].astype(str)
```

```python
#change date format
for i, row in state_df.iterrows():
    parsed= parse(row.date)
    state_df.at[i,'date']=parsed.strftime('%Y-%m-%d')
```

```python
us_df=us_df.drop(columns=['states'])
```

```python
us_df.drop(us_df.iloc[:,2:25],inplace=True,axis=1)
```

```python
us_df.rename(columns={'date':'Date','positive':'NationalPositiveCount'},inplace=True)
us_df.head()
```

```python
state_df.head()
```

```python
state_df.drop(state_df.iloc[:,3:41],inplace=True,axis=1)
state_df.drop(state_df.iloc[:,4:18],inplace=True,axis=1)
state_df.rename(columns={'date':'Date','state':'State','positive':'TotalStateCount','positiveIncrease':'DailyIncrease'},inplace=True)
state_df = state_df[['Date','State','DailyIncrease','TotalStateCount']]
state_df.head()
```

```python
merged_df=pd.merge(us_df,state_df,on='Date')
merged_df = merged_df[['Date','State','DailyIncrease','TotalStateCount','NationalPositiveCount']]
merged_df.head(50)
```

```python
merged_df.to_csv('data/cleanusa.csv')
```
